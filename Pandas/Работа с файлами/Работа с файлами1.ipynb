```python
df = pd.read_csv(r'C:\Users\user\Desktop\Skilfactory\data\recipes.csv')
```

```python
df.drop('ingredients', inplace=True, axis=1)
```

```python
ids = []
for i in df['id']:
    ids.append(i)
```

```python
df.head()
```

```python
# Создаём пустой список для хранения итоговой структуры
new_recipes = [] 

# Получаем список всех ингредиентов (столбцов)
ingredients = list(df.columns)[3:]  # Предполагаем, что первые 3 столбца - не ингредиенты

def make_list(row):
    """Функция для получения списка ингредиентов блюда"""
    return [ing for ing in ingredients if row[ing].iloc[0] == 1]

# Организуем цикл по всем ID
for current_id in ids: 
    # Фильтруем DataFrame по текущему ID
    current_df = df[df['id'] == current_id]
    
    # Получаем значение кухни
    cuisine = current_df['cuisine'].iloc[0] 
    
    # Получаем перечень ингредиентов
    current_ingredients = make_list(current_df)
    
    # Создаём текущий словарь
    current_recipe = {
        'cuisine': cuisine, 
        'id': int(current_id), 
        'ingredients': current_ingredients
    } 
    
    # Добавляем созданный словарь к списку
    new_recipes.append(current_recipe)
```

```python
import json 
# Функция dumps() модуля json сериализирует объект Python в строку формата JSON. 
new_recipes = json.dumps(new_recipes) 

# Откроем файл new_recipes.json для записи
with open(r"C:\Users\user\Desktop\Skilfactory\data/new_recipes.json", "w") as write_file: 
    # Записываем содержимое подготовленные данные в файл
    write_file.write(new_recipes)
```
