```python
vaccinations_data = vaccinations_data[
    ['country', 'date', 'total_vaccinations',
     'people_vaccinated', 'people_vaccinated_per_hundred',
     'people_fully_vaccinated', 'people_fully_vaccinated_per_hundred',
     'daily_vaccinations', 'vaccines']
]
```

```python
covid_data = covid_data.groupby(
    ['date', 'country'],
    as_index=False
)[['confirmed', 'deaths', 'recovered']].sum()
```

```python
covid_data['date'] = pd.to_datetime(covid_data['date'])
```

```python
covid_data['active'] = covid_data['confirmed'] - covid_data['deaths'] - covid_data['recovered']
```

```python
covid_data = covid_data.sort_values(by=['country', 'date'])
covid_data['daily_confirmed'] = covid_data.groupby('country')['confirmed'].diff()
covid_data['daily_deaths'] = covid_data.groupby('country')['deaths'].diff()
covid_data['daily_recovered'] = covid_data.groupby('country')['recovered'].diff()
```

```python
import plotly.express as px
import os
from pathlib import Path

# 1. Подготовка пути для сохранения
plots_dir = Path(r"C:\Users\user\Desktop\Skilfactory\plots")
plots_dir.mkdir(parents=True, exist_ok=True)  # Создаем папку если не существует
output_path = plots_dir / "covid_plot.svg"

# 2. Обработка данных
line_data = covid_df.groupby('date', as_index=False).sum()

# 3. Построение графика
fig = px.line(
    data_frame=line_data,
    x='date',
    y=['confirmed', 'recovered', 'deaths', 'active'],
    height=500,
    width=1000,
    title='COVID-19 Cases Over Time',
    labels={
        'date': 'Date',
        'value': 'Number of Cases',
        'variable': 'Case Type'
    }
)

# 4. Дополнительные настройки
fig.update_layout(
    hovermode="x unified",
    legend_title_text='Case Type',
    xaxis_title='Date',
    yaxis_title='Number of Cases'
)

# 5. Сохранение и отображение с обработкой ошибок
try:
    fig.write_image(output_path, engine="kaleido")
    print(f"График успешно сохранен: {output_path}")
except Exception as e:
    print(f"Ошибка при сохранении графика: {e}")
    try:
        fig.write_image(output_path.with_suffix('.png'))
        print("Сохранено как PNG вместо SVG")
    except Exception as e:
        print(f"Не удалось сохранить график: {e}")

# 6. Отображение графика
fig.show()
```
